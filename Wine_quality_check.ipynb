```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd

# Assuming you have your DataFrame df with the columns 'fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar','chlorides', 'free_ sulfur_ dioxide', 'total_sulfur_dioxide', 'density','pH', 'sulphates', 'alcohol'

df = pd.read_csv('/content/Winequality Red.csv')
# Extract features (X) and target variable (y)
X = df[['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar','chlorides', 'free_ sulfur_ dioxide', 'total_sulfur_dioxide', 'density','pH', 'sulphates', 'alcohol']]
y = df['quality']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a linear regression model
model = LinearRegression()

# Train the model using the training sets
model.fit(X_train, y_train)

# Now, suppose you have new data for which you want to predict wine quality
# Prompt the user to enter the 'fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar','chlorides', 'free_ sulfur_ dioxide', 'total_sulfur_dioxide', 'density','pH', 'sulphates', 'alcohol'

print("Enter the acidity of the wine:")

# Assuming your wine quality
n1 = float(input("Enter the fixed_acidity: "))
n2 = float(input("Enter the volatile_acidity: "))
n3 = float(input("Enter the citric_acid: "))
n4 = float(input("Enter the residual_sugar: "))
n5 = float(input("Enter the chlorides: "))
n6 = float(input("Enter the free_sulfur_dioxide: "))
n7 = float(input("Enter the total_sulfur_dioxide: "))
n8 = float(input("Enter the density: "))
n9 = float(input("Enter the pH: "))
n10 = float(input("Enter the sulphates: "))
n11 = float(input("Enter the alcohol: "))

# Create a DataFrame for the new data
new_data = pd.DataFrame([[n1,n2,n3,n4,n5,n6,n7,n8,n9,n10,n11]], columns=['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar','chlorides', 'free_ sulfur_ dioxide', 'total_sulfur_dioxide', 'density', 'pH', 'sulphates', 'alcohol'])

# Make prediction
prediction = model.predict(new_data)

# Print the predicted wine quality
print("Predicted wine quality:", prediction[0])


```
